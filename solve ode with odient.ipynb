```python
from scipy.integrate import  odeint
import numpy as np 
import matplotlib.pyplot as plt 
def model(y,t):
    k= 0.3
    dydt= -k*y**2
    return dydt
y0=5
t=np.linspace(0,20)
y=odeint(model ,  y0 , t)

plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)') 
plt.show()   
```

```python
def model(y, t, k):
    dydt = -k * y**2
    return dydt

k = 0.1
y1 = odeint(model, y0, t, args=(k,))
k = 0.2
y2 = odeint(model, y0, t, args=(k,))

plt.plot(t, y1, label='k1')
plt.plot(t, y2,label='k2')
plt.legend()
plt.show()
```
